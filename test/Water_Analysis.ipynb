# Top title

The notebook headers will be printed during testing whenever the verbosity level `nbtest(...; verbose=3)` is greater than or equal to the number of `#` of the header. By default, all headers are printed (`verbose=5`)

```julia
using NBTesting
using Base.Test

# This defines 
#    plot(args...; kwargs...) = nothing
# but only during testing.
@testing_noop plot
#NBSKIP
using Plots
```

```julia
x = 10 - 9
@test x==1
```

### Basic Arithmetic

The last test fails, but because it's after `#NBSKIP`, it won't be executed during testing. `#NBSKIP` is also useful for skipping long computations.

```julia
@test 5+5 == 10
@test 6*10 == 60
#NBSKIP
@test 7-7 == 3
```

```julia
# More complicated code and test
srand(1)
M1 = ones(5,5)
M2 = rand(5,5);
```

```julia
@test 30 < sum(M1 * M2) < 100
```
